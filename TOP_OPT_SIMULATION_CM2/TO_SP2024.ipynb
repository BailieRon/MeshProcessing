```python
from fenics import *
import matplotlib.pyplot as plt

# Define the geometry parameters
Lx = 2.0  # Length of the plate in the x-direction
Ly = 1.0  # Length of the plate in the y-direction

# Define the mesh
nx, ny = 50, 25  # Number of elements in the x and y directions
mesh = RectangleMesh(Point(0, 0), Point(Lx, Ly), nx, ny)

# Define the function space for density
V = FunctionSpace(mesh, 'P', 1)

# Define the boundary conditions
def boundary(x, on_boundary):
    return on_boundary

bc = [DirichletBC(V, Constant(1.0), 'near(x[0], 0)'),
      DirichletBC(V, Constant(0.0), 'near(x[0], Lx)')]

# Define the density variable
rho = Function(V)
v = TestFunction(V)

# Define the optimization problem
F = inner(grad(v), grad(rho))*dx + 1e-3*rho*v*dx
problem = VariationalProblem(F, derivative(F, rho), bc)
solver = VariationalSolver(problem)

# Set up the optimization parameters
parameters = solver.parameters
parameters['nonlinear_solver'] = 'snes'
parameters['snes_solver']['line_search'] = 'bt'
parameters['snes_solver']['maximum_iterations'] = 100

# Solve the optimization problem
solver.solve()

# Plot the optimized density
plot(rho)
plt.show()
```
